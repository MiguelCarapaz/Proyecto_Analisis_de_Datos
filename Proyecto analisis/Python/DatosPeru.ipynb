```python
df_peru = pd.read_csv('D:\descargas\Matriculados_2016_al_2022.csv', delimiter=",")
df_peru
```

```python
# Crear una lista de valores para la nueva columna
valores_causas = list(range(1, 11)) + [None]*(len(df_peru)-10)

# Agregar la nueva columna al dataframe
df_peru['Causas'] = valores_causas

# Seleciona los datos entre 2018 y 2020
busqueda = df_peru.loc[df_peru['Año Matrícula'] >= 2018]
busqueda = busqueda.loc[df_peru['Año Matrícula'] <= 2021]

# Comenzar el indice en cero
busqueda = busqueda.reset_index(drop=True)

# Mostrar los datos selecionados
busqueda.isnull().sum()
```

```python
busqueda["Género"].fillna(value="N/A", inplace=True)
busqueda.isnull().sum()
#busqueda
```

```python
busqueda.loc[0:3, 'Causas'] = 'Factores Socioeconónicos'
busqueda.loc[4:100, 'Causas'] = 'Factores Institucionales'
busqueda.loc[101:2000, 'Causas'] = 'Factores Familiares'
busqueda.loc[2001:4000, 'Causas'] = 'Factores Sociales'
busqueda.loc[4001:6000, 'Causas'] = 'Factores Familiares'
busqueda.loc[6001:8000, 'Causas'] = 'Factores Personales'
busqueda.loc[8001:10000, 'Causas'] = 'Factores Psicológicos'
busqueda.loc[10001:23000, 'Causas'] = 'Factores Socioeconómicos'
busqueda.loc[23001:40000, 'Causas'] = 'Factores Psicológicos'
busqueda.loc[40001:60000, 'Causas'] = 'Factores Socioeconómicos'
busqueda.loc[60001:70818, 'Causas'] = 'Factores Familiares'
busqueda.loc[70819:93452, 'Causas'] = 'Factores Socioeconómicos'
busqueda
```

```python
direccion_servidor = 'localhost'
nombre_bd = 'proyecto'
nombre_usuario = 'root'
password = 'Ligacampeon1'
```

```python
try:
    # Conectar a la base de datos
    conexion = mysql.connector.connect(
        host=direccion_servidor,
        database=nombre_bd,
        user=nombre_usuario,
        password=password
    )

    # Obtener un cursor para interactuar con la base de datos
    cursor = conexion.cursor()

    # Iterar sobre el DataFrame e insertar los datos en la tabla
    for index, row in busqueda.iterrows():
        cursor.execute("INSERT INTO Datos_Peru (Anio_Matricula,Periodo,Tipo_hecho,Genero,Causas) VALUES (%s,%s,%s,%s,%s)", (row['Año Matrícula'], row['Período'], row['Tipo de hecho'], row['Género'], row['Causas']))

    # Confirmar los cambios
    conexion.commit()
    print("Datos insertados correctamente")

except Exception as e:
    # Atrapar error
    print("Ocurrió un error al insertar los datos: ", e)
```

```python
query = "SELECT * FROM Datos_Peru"
datos_peru= pd.read_sql_query(query, conexion)
datos_peru
```
