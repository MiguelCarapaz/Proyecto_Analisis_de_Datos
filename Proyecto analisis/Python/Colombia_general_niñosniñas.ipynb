```python
couch = couchdb.Server('http://miguel:miguel@localhost:5984/')
db = couch['colombia']
```

```python
import pandas as pd

data = pd.read_csv('Colombia.csv')
docs = data.to_dict(orient='records')
```

```python
for doc in docs:
    for key, value in doc.items():
        if isinstance(value, float):
            doc[key] = str(value)
    db.save(doc)
```

```python
import json

for doc in docs:
    json_doc = json.dumps(doc, allow_nan=True)
    db.save(json.loads(json_doc))
```

```python
#SUBIR A MONGODB los datos de Ecuador
from pymongo import MongoClient
# Conectarse al servidor local de MongoDB
client = MongoClient('localhost', 27017)
# Seleccionar una base de datos
db = client['mydatabase']
# Seleccionar una colección
collection = db['datos_generales']
```

```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('datos_generales.csv')

# Convertir los datos a formato JSON
records = df.to_dict(orient='records')
```

```python
# Insertar los datos en la colección
result = collection.insert_many(records)

# Imprimir el número de documentos insertados
print('Se insertaron {} documentos en la colección'.format(len(result.inserted_ids)))
```

```python
# Obtener todos los documentos de la colección
documents = collection.find()

# Crear una lista de diccionarios a partir de los documentos
docs_list = [doc for doc in documents]

# Crear un DataFrame de pandas a partir de la lista de diccionarios
df3 = pd.DataFrame(docs_list)

# Imprimir el DataFrame
df3
```
